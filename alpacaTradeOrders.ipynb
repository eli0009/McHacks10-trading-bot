# Imports

```python
import json

# Importing the API 
from alpaca.trading.client import TradingClient

# Importing the necessary apis for making a buy order
from alpaca.trading.requests import MarketOrderRequest
from alpaca.trading.enums import OrderSide, TimeInForce
```

# Creating Trading Client

```python
# Instantiating the REST client according to the keys
with open(credentialsPath, 'r') as jsonfile:
    data = json.load(jsonfile)
    USERNAME = data["username"]
    PASSWORD = data["password"]
    API_KEY = data["api_key"]
    API_SECRET = data["api_secret"]

#Creating a paper trading client
trading_client = TradingClient(API_KEY, API_SECRET, paper=True)
```

Buying power can be found inside oneâ€™s account information. Use the get_account method on the trading client and print the result to show your account information.

# To get historic crypto data

```python
from alpaca.data.historical import CryptoHistoricalDataClient

# No keys required for crypto data
client = CryptoHistoricalDataClient()
```

```python
from alpaca.data.requests import CryptoBarsRequest
from alpaca.data.timeframe import TimeFrame

# Creating request object
request_params = CryptoBarsRequest(
                        symbol_or_symbols=["BTC/USD"],
                        timeframe=TimeFrame.Day,
                        # start="2022-09-01",
                        # end="2022-09-07"
                        )
print(request_params)
```

```python
# Retrieve daily bars for Bitcoin in a DataFrame and printing it
btc_bars = client.get_crypto_bars(request_params)

# Convert to dataframe
btc_bars.df
```

# Streaming Real-Time Data

```python
from alpaca.data.live import CryptoDataStream
```

```python
crypto_stream = CryptoDataStream(API_KEY, API_SECRET)
```

```python
async def bar_callback(bar):
    for property_name, value in bar:
        print(f"\"{property_name}\": {value}")

# Subscribing to bar event 
symbol = "BTC/USD"
crypto_stream.subscribe_bars(bar_callback, symbol)

crypto_stream.run()
```

```python
import requests

# Replace <KEY> and <SECRET> with your actual API key and secret
headers = {
    'APCA-API-KEY-ID': API_KEY,
    'APCA-API-SECRET-KEY': API_SECRET
}

# date format is 'YYYY-MM-DD'
date = '2022-01-01'
url = f'https://data.alpaca.markets/v1beta1/news?start_date={date}&end_date={date}'

response = requests.get(url, headers=headers)

# Check the status of the request
if response.status_code == 200:
    # Print the historical news of the specified date
    print(response.json())
else:
    print(f'Error: {response.status_code}')
```
